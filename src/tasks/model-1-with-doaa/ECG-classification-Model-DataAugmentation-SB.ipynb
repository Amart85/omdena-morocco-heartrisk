```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("C:/Users/TOSHIBA/Desktop/mitbih_train.csv" , header = None)
test_data = pd.read_csv("C:/Users/TOSHIBA/Desktop/mitbih_test.csv" , header = None)
```

```python
train_data
```

```python
C_1 = train_data[train_data[187] == 1]
C_2 = train_data[train_data[187] == 2]
C_3 = train_data[train_data[187] == 3]
C_4 = train_data[train_data[187] == 4]
```

```python
from sklearn.utils import resample

C_1_upsample = resample(C_1, n_samples = 20000, replace = True, random_state = 123)
C_2_upsample = resample(C_2, n_samples = 20000, replace = True, random_state = 123)
C_3_upsample = resample(C_3, n_samples = 20000, replace = True, random_state = 123)
C_4_upsample = resample(C_4, n_samples = 20000, replace = True, random_state = 123)
```

```python
C_0 = train_data[train_data[187]==0].sample(n =20000, random_state=123)
```

```python
train_C = pd.concat([C_0, C_1_upsample, C_2_upsample, C_3_upsample, C_4_upsample])
```

```python
target_train = train_C[187]
target_test = test_data[187]
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(target_train)
```

```python
y_test = to_categorical(target_test)
```

```python
X_train = train_C.iloc[:, :-1].values
X_test = test_data.iloc[:, :-1].values
```

```python
X_train = X_train.reshape(len(X_train),X_train.shape[1],1)
X_test = X_test.reshape(len(X_test),X_test.shape[1],1)
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense 
from keras.layers import Conv1D, MaxPooling1D, Flatten
from keras.optimizers import Adam


from keras.layers.normalization import BatchNormalization
```

```python
def create_model():
    model = Sequential()
    
    
    model.add(Conv1D(filters = 64, kernel_size = 6, activation='relu', padding = 'same', input_shape = (187, 1))) 
    
  
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=(3), strides = (2), padding = 'same'))
    model.add(Conv1D(filters = 64, kernel_size = 6, activation='relu', padding = 'same'))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=(3), strides = (2), padding = 'same'))
    model.add(Conv1D( filters = 64, kernel_size = 6, activation='relu', padding = 'same'))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=(3), strides = (2), padding = 'same'))
    model.add(Flatten())
    model.add(Dense(units = 64, activation='relu'))
    model.add(Dense(units = 64, activation='relu'))
    model.add(Dense(units = 5, activation='softmax'))
    model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
model = create_model()
```

```python
model.summary()
```

```python
F = model.fit(X_train, y_train, epochs = 15, batch_size = 32, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```
